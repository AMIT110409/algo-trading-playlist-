```python
import pandas as pd
```

```python
df = pd.read_excel('sample.xlsx')
```

```python
df.to_csv('sample.csv', index=False)
```

```python
df.head()
```

```python
columns_with_nan = df.columns[df.isna().any()]
```

```python
df.drop(columns_with_nan,axis=1,inplace=True)
```

```python
df.head(5)
```

```python
# Calculate net brokerage
df['Net Brokerage'] = df['Buy Brokerage'] + df['Sell Brokerage']
```

```python
# Calculate net quantity traded
df['Total Qty'] = df['Buy Qty'] + df['Sell Qty']
```

```python
df.head()
```

```python
## here i am writing code for what is total quantity traded[buy& sell] by the client S?
```

```python
total_quantity_traded_by_s = df[df['ClientCode'] == 'S']['Total Qty'].sum()
```

```python
print(print("Total Quantity Traded by Client 'S':", total_quantity_traded_by_s)
)
```

```python
df['Expiry Date'] = pd.to_datetime(df['Expiry Date'])
# convert the expiry date into  datetime formet .
```

```python
# # Calculate brokerage paid by client 'M' in expiry week 20210128
# brokerage_paid_by_m_in_20210128 = df[(df['ClientCode'] == 'M') & (df['Expiry Date'] == '20210128')]['Net Brokerage'].sum()
# print("Brokerage Paid by Client 'M' in Expiry Week 20210128:", brokerage_paid_by_m_in_20210128)


# Brokerage Paid by Client 'M' in the Expiry Week 20210128
brokerage_paid_by_m_in_week_20210128 = df[(df['ClientCode'] == 'M') & (df['Expiry Date'] >= '2021-01-23') & (df['Expiry Date'] <= '2021-01-29')]['Net Brokerage'].sum()
print("Brokerage Paid by Client 'M' in the Expiry Week 20210128:", brokerage_paid_by_m_in_week_20210128)
```
